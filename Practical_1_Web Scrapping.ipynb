```python
#import requests: This line imports the requests module, which you installed in the first line. It provides functions for making HTTP requests
import requests
```

```python
#This line imports the BeautifulSoup class from the bs4 (Beautiful Soup 4) module. BeautifulSoup is used for pulling data out of HTML and XML files.
from bs4 import BeautifulSoup
```

```python
#This line sends an HTTP GET request to the specified URL using the requests.get function and stores the response in the variable req.
req = requests.get("https://www.rjcollege.edu.in")
soup = BeautifulSoup(req.content, "html.parser")
```

```python
soup = BeautifulSoup(req.content, "html.parser")
print(soup.prettify())
```

```python
#This prints the entire text content of the HTML using BeautifulSoup's get_text() method.
print(soup.get_text())
```

```python
#This finds the title tag in the HTML, stores it in title_tag, and then prints the prettified HTML of the title tag.
title_tag = soup.title
print("Prettified Title Tag:")
print(title_tag.prettify())
```

```python
#This prints the text content of the title tag using get_text().
print("Title Tag Text:", title_tag.get_text())
```

```python
#This finds all anchor tags (<a>) in the HTML, iterates through them, and prints their href attributes and text content.
links = soup.find_all("a")
for link in links:
    print("Link:", link.get("href"), "Text:", link.string)
```
